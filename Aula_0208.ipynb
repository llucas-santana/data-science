4. Durante uma corrida de automóveis com N voltas de duração foram anotados para um piloto, na ordem, os tempos registrados em cada volta. Fazer um programa para ler os tempos das N voltas, calcular e imprimir: melhor tempo; a volta em que o melhor tempo ocorreu; tempo médio das N voltas.

```python
quantidadeVoltas = int(input("Informe quantidade de voltas: "))
tempoVoltaMaisRapida = 99999.0
voltaMaisRapida = 0
tempoMedio = 0

for volta in range(1, quantidadeVoltas + 1, 1):
  tempoVolta = float(input(f"Informe o tempo da volta [{volta}]: "))
  if tempoVolta < tempoVoltaMaisRapida:
    tempoVoltaMaisRapida = tempoVolta
    voltaMaisRapida = volta
  tempoMedio += tempoVolta / quantidadeVoltas

print(f"Tempo Medio: {tempoMedio}")

print(f"A volta {voltaMaisRapida} fez o melhor tempo de {tempoVoltaMaisRapida}")

```

6. Desenvolva um programa que receba quatro notas de um aluno, calcule e imprima a média aritmética das notas e a mensagem de aprovado para média superior ou igual a 7.0 ou a mensagem de reprovado para média inferior a 7.0.

```python
total = 0.0

for i in range(0, 4, 1):
  nota = float(input(f"Informe a nota[{i}]: "))
  total += nota

media = total / 4;

if media >= 7.0:
  print("Aprovado")
else:
  print("Reprovado")

print(f"Media eh: {media}")
```

7. Escreva um programa para validar um lote de cheques. O programa deverá inicialmente solicitar a soma do lote e o número de cheques. A seguir deverá ler o valor de cada cheque calculando a soma total. Após a digitação de todos os cheques o programa deverá imprimir as seguintes mensagens:

  a. LOTE Ok se a soma informada for igual a soma calculada.
  
  b. Diferença negativa se a soma calculada for menor que a informada.


```python
somaLote = float(input("Informe a soma do lote: "))
numeroCheques = int(input("Numero de cheques: "))
somaDosCheques = 0
diferenca = 0

for i in range(0, numeroCheques, 1):
  valorCheque = float(input("Informe o valor do cheque: "))
  somaDosCheques += valorCheque

diferenca = somaDosCheques - somaLote

if somaDosCheques == somaLote:
  print("LOTE OK")
elif somaDosCheques < somaLote: 
  print(f"Diferença negativa {diferenca}")
else: 
  print(f"Diferença positiva {diferenca}")
```
